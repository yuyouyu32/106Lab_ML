```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
result_URL = "/home/yuyouyu/WorkSpace/106Lab_Data/fatigue_prediction/9-16/result.json"
with open(result_URL,'r',encoding='utf8')as f:
    result = json.load(f)

target_names = {'屈服强度': 'yield_strength',
               '延伸率': 'elongation',
               '抗拉强度': 'Tensile_strength',
               '断面收缩率':'Reduction_of_section'}
x = range(1, 11)
scores = result["NN"]["R2"]
plt.figure(figsize=(25, 25))
for index, target_name in enumerate(scores):
    plt.ylim(0, 1.2)
    pl = plt.subplot(2, 2, index + 1)
    pl.set_yticks(np.linspace(0, 1, 10))
    pl.plot(x, scores[target_name]["Train"],label="Train",color="Blue", linestyle="dashed",
            marker='o')
    pl.plot(x, scores[target_name]["Test"],label="Test",color="Red", linestyle="dashdot",
           marker='*')
    pl.grid(alpha=0.5,linestyle=':')
    average = sum(scores[target_name]['Test'])/ 10.0
    name = target_names[target_name]
    pl.set_title(f"{name}: {average}", )
```
