```python
import pandas as pd

data = pd.read_csv('点蚀电位清洗版.csv')
data = data.fillna(0)
values = data.astype('double')
out_result = values.loc[:, (values != 0).any(axis=0)]
out_result.to_csv('点蚀电位清洗版-1.csv')
```

```python
data = pd.read_csv('PREN.csv')
data = data.fillna(0)
print(data.describe())
values = data.astype('double')
out_result = values.loc[:, (values != 0).any(axis=0)]
out_result.to_csv('PREN-1.csv')
print(out_result.describe())
```
