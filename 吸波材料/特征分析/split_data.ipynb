```python
import pandas as pd
```

# 最强吸收值

```python
def process(input_str: str):
    return list(map(float, input_str.split(',')))

original_data = pd.read_csv('./除定石蜡比吸收值.csv')
out_data = pd.DataFrame(dtype='float64')
    
choose_features = ['Hz', 'u', 'widths', 'width_heights', 'prominences']
for index, row in original_data.iterrows():
    Hzs = process(row['Hz'])
    peak_num = len(Hzs)
    if peak_num != 3:
        continue
    temp = pd.Series(dtype='float64')
    for feature in choose_features:
        temp_features = process(row[feature])
        temp[f'{feature}_1'], temp[f'{feature}_2'], temp[f'{feature}_3'] = temp_features
    temp['minRL'] = row['minRL']
    out_data = out_data.append(temp, ignore_index=True)
out_data
```

```python
out_data.to_csv('minRL.csv', index=False)
```
