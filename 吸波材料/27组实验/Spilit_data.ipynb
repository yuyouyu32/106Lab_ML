```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('./27_data.csv')
target_names = ['d_of_Bandwidth','Bandwidth','d_of_minRL','F_of_minRL','minRL']
data
```

## All Data

```python
for target_name in target_names:
    target = data.loc[:, target_name]
    features = data.drop(columns=target_names)
    split_data = pd.concat([target, features],axis=1)
    split_data.to_csv(f'{target_name}.dat', sep=' ', index=True)
```

## Train and Test

```python
for target_name in target_names:
    target = data.loc[:, target_name]
    features = data.drop(columns=target_names)
    Xtrain, Xtest, Ytrain, Ytest = train_test_split(features,target,test_size=0.3,random_state=32)
    train_data = pd.concat([Ytrain, Xtrain],axis=1)
    train_data.to_csv(f'{target_name}_train.dat', sep=' ', index=True)
    test_data = pd.concat([Ytest, Xtest],axis=1)
    test_data.to_csv(f'{target_name}_test.dat', sep=' ', index=True)
```
