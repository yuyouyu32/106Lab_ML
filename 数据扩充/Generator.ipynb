```python
import random
import numpy as np
import decimal
def generater_list(start: float, end: float, step: float, if_random: bool = False):
    generation = list(np.arange(start, end + step, step))
    d_num = abs(decimal.Decimal(str(step)).as_tuple().exponent)
    if if_random:
        for i in range(len(generation)):
            random_num = random.uniform(step/10, step)
            generation[i] = generation[i] +random_num
    return [round(temp, d_num + 1) for temp in generation]
```

# 7X

```python
from itertools import product
import pandas as pd

Original_data = pd.read_csv('./7X/train.csv', encoding='gbk')
target_columns = ['维氏硬度（HV）',	'维氏硬度（HV）方差','屈服强度（Mpa）',	'屈服强度（Mpa）方差', '抗拉强度（Mpa）', '抗拉强度（Mpa）方差', '延伸率（%）', '延伸率（%）方差']
colnames = list(Original_data.drop(columns=target_columns).columns)
```

```python
C_1 = [100, 85, 84.5, 84, 84.5 ]
C_2 = [0, 0.5, 1, 1.5]
C_3 = [0, 14.5, 15]
C_4 = [0, 0.5]
T_1 = generater_list(460, 475, 1)
T_2 = generater_list(30, 60, 1)
T_3 = [120]
T_4 = generater_list(9, 24, 1)
data = list(product(C_1, C_2, C_3, C_4, T_1, T_2, T_3, T_4))
data = pd.DataFrame(data , columns=colnames)
data.drop(data[(data['Al91Zn5.2Mg2.3Cu1.5'] + data['Mg'] + data['SiCp'] + data['CNTs']!=100)].index, inplace=True)
data.drop(data[(data['Al91Zn5.2Mg2.3Cu1.5'] + data['Mg'] != 85) & (data['SiCp'] + data['CNTs'] > 0)].index, inplace=True)
```

```python
data.to_csv('./7X/ex_data.csv', index=False)
```

```python
data_shuffle = data.sample(frac = 1)
data_shuffle.to_csv('./7X/ex_data_shuffle.csv', index=False)
```

```python
from itertools import product
import pandas as pd

Original_data = pd.read_csv('./负膨胀/train.csv', encoding='gbk')
Original_data.fillna(0, inplace=True)
target_columns = ['neg_inflation', 's_temp', 'e_temp', 'temp_range', 'formula']
colnames = list(Original_data.drop(columns=target_columns).columns)
```

```python
Mn = list(set(generater_list(2.6, 3, 0.02) + generater_list(2.6, 3, 0.05)))
Fe = generater_list(0, 0.12, 0.02)
Cr = generater_list(0,0.35, 0.05)
Sn = generater_list(0,0.05, 0.05)
part_1 = list(product(Mn, Fe, Cr, Sn))
part_1 = pd.DataFrame(part_1 , columns=colnames[0:4])
part_1.drop(part_1[(part_1['Mn'] + part_1['Fe'] + part_1['Cr'] + part_1['Sn']!=3)].index, inplace=True)
len(part_1)
```
